```python
import pandas as pd
import numpy as np
from keras.datasets import mnist
import gradio as gr
import tensorflow as tf
from tensorflow import keras 
from keras import Sequential
from keras.layers import Dense,Conv2D,Flatten,MaxPool2D,Dropout
from tensorflow.keras.callbacks import EarlyStopping as ES
import matplotlib.pyplot as mp
from sklearn.preprocessing import StandardScaler as SS
from keras.models import Model
from sklearn.metrics import accuracy_score as AS
```

```python
(xta, yta), (xte, yte) = mnist.load_data()
```

```python
xta.shape
```

```python
f,a=mp.subplots(4,4,figsize=(10,8))
for i in range(4):
    for j in range(4):
        a[i][j].imshow(xta[i+j].reshape(28,28))
```

```python
xta=xta/255
xte=xte/255
```

```python
model=Sequential()
model.add(Conv2D(20,kernel_size=(3,3),padding="same",activation="relu",input_shape=(28,28,1)))
model.add(MaxPool2D(pool_size=(2, 2),strides=2))

model.add(Conv2D(40,kernel_size=(3,3),padding="same",activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2),strides=2))

model.add(Flatten())
model.add(Dense(20,activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(10,activation="softmax"))


model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(momentum=0.8,nesterov=True),loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
es=ES(monitor="val_loss",
    min_delta=0.00001,
    patience=10,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0,)
```

```python
model.fit(xta,yta,validation_data=(xte,yte),epochs=30,batch_size=15,callbacks=es)
```
